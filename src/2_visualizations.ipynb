### Wizualizacje – Dirty Cafe Sales
Przygotowany notatnik Jupyter zgodny z wymaganiami (4-5 wykresów + opis).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Wczytanie danych
df = pd.read_csv('clean_cafe_sales.csv')

# Ustawienie stylu wykresów
sns.set(style="whitegrid")
```

### Opis wykresu
Ten wykres pokazuje, które produkty cieszą się największą popularnością w kawiarni.

```python
# 1. Najpopularniejsze produkty (Item)
plt.figure(figsize=(8,5))
df['Item'].value_counts().plot(kind='bar')
plt.title('Najczęściej kupowane produkty')
plt.ylabel('Liczba transakcji')
plt.xlabel('Produkt')
plt.tight_layout()
plt.show()
```

### Opis wykresu
Histogram przedstawia rozkład kwot wydawanych podczas pojedynczej transakcji oraz wskazuje typowy zakres wydatków.

```python
# 2. Rozkład wydatków na transakcję (Total Spent)
plt.figure(figsize=(8,5))
sns.histplot(df['Total Spent'], bins=20, kde=True)
plt.title('Rozkład kwot wydanych na transakcję')
plt.xlabel('Kwota wydana [USD]')
plt.ylabel('Liczba transakcji')
plt.tight_layout()
plt.show()
```

### Opis wykresu
Wykres pokazuje preferencje klientów dotyczące metod płatności w zależności od tego, czy kupują na miejscu, czy na wynos.

```python
# 3. Metody płatności w zależności od lokalizacji
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Payment Method', hue='Location')
plt.title('Metody płatności w zależności od lokalizacji')
plt.xlabel('Metoda płatności')
plt.ylabel('Liczba transakcji')
plt.legend(title='Lokalizacja')
plt.tight_layout()
plt.show()
```

### Opis wykresu
Ten wykres przedstawia, w które dni miesiąca klienci wydają średnio najwięcej pieniędzy.

```python
# 4. Średnia sprzedaż w poszczególne dni miesiąca
plt.figure(figsize=(10,6))
df_grouped = df.groupby('Day')['Total Spent'].mean()
df_grouped.plot(kind='bar')
plt.title('Średnia kwota wydana w poszczególne dni miesiąca')
plt.ylabel('Średnia kwota wydana [USD]')
plt.xlabel('Dzień miesiąca')
plt.tight_layout()
plt.show()
```

### Opis wykresu

Ten wykres pudełkowy przedstawia rozrzut kwot wydanych podczas transakcji. Pozwala zaobserwować medianę, kwartyle oraz wartości odstające (outliery), jeśli występują.

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Total Spent'])
plt.title('Rozrzut kwot wydanych na transakcję (Boxplot)')
plt.xlabel('Kwota wydana [USD]')
plt.tight_layout()
plt.show()
```
